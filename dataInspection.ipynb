```python
extraData=pd.read_excel("Data_Train.xlsx").drop(columns=["Additional_Info","Route"])
extraData.dropna(inplace=True)
extraData.head()
```

```python
extraData["Destination"].unique() ,extraData["Source"].unique()
```

```python
def strToHours(x):
    if type(x)==str:
        if "h" in x:
            if "m" in x:
                return int(x.split("h")[0])*60+int(x.split("h")[1].split("m")[0])
            else:
                return int(x.split("h")[0])*60
        else:
            return int(x.split("m")[0])
    else:
        return x
strToHours("2h 50m"), strToHours("50m"), strToHours(50)
```

```python
def stopCount(x):
    if type(x)==str:
        split=x.split(" ")[0]
        if split=="non-stop":
            return 0
        else:
            return int(split)
    else:
        return x
```

```python
extraData["Duration"]=extraData["Duration"].apply(lambda x: strToHours(x))
extraData["Total_Stops"]=extraData["Total_Stops"].apply(lambda x: stopCount(x))
extraData["Date_of_Journey"] = pd.to_datetime(extraData.Date_of_Journey, format="%d/%m/%Y")
```
